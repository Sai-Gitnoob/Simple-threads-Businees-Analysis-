```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("D:\Coding\College IITJ\BDA\Capstone\Customer_Information (cleaned).csv")
```

```python
df["Address"].value_counts().plot(kind="bar")
plt.title("Customer Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.savefig("customer_distribution_by_region.png", dpi=300, bbox_inches="tight")
plt.show()
```
